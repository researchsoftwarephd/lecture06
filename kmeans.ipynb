```python
from sklearn.datasets import make_blobs      # Function to create data
from sklearn.cluster import KMeans           # Class to perform k-Means clustering
from sklearn.metrics import v_measure_score  # Function to assess clustering quality
import matplotlib.pyplot as plt              # Module to plot stuff
```

```python
xy, groups = make_blobs(1000, 2, centers=4, cluster_std=2.0, random_state=123) # Generate some data...
plt.scatter(xy[:, 0], xy[:, 1], c=groups)                    # ...and show it
```

```python
result = KMeans(n_clusters=4, random_state=432).fit(xy)              # Automatically group data with k-means
plt.scatter(xy[:, 0], xy[:, 1], c=result.labels_)  # Show how k-means grouped our data
```

```python
v_measure_score(groups, result.labels_)  # Get clustering quality, between 0 and 1
```
